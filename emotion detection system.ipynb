```python
import tkinter as tk
from tkinter import Label
import cv2
from PIL import Image, ImageTk
from deepface import DeepFace
import pyttsx3
import threading

engine = pyttsx3.init() # Initialize text-to-speech engine

def detect_emotion(frame):# Function to detect emotion and update the label
    try:
        result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
        if isinstance(result, list) and len(result) > 0:
            # Multiple faces, use the first one
            emotion = result[0]['dominant_emotion']
        else:
            # Single face detected or normal result
            emotion = result['dominant_emotion']
        return emotion
    except Exception as e:
        print(f"Emotion detection error: {e}")
        return None

# Function to handle the text-to-speech output
def speak_emotion(emotion):
    if emotion:
        engine.say(f"The person looks {emotion}.")
        engine.runAndWait()

# Function to update the webcam feed in the GUI
def update_frame():
    ret, frame = cap.read()
    if ret:
        # Convert frame to RGB (since OpenCV uses BGR)
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Detect emotion
        emotion = detect_emotion(frame)

        # Draw emotion on the frame
        if emotion:
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(frame, f'Emotion: {emotion}', (10, 30), font, 1, (0, 255, 0), 2, cv2.LINE_AA)

        # Convert the frame to a format Tkinter can display
        img = Image.fromarray(frame)
        imgtk = ImageTk.PhotoImage(image=img)

        # Update the label with the new image
        lbl_video.imgtk = imgtk
        lbl_video.configure(image=imgtk)

        # Update the emotion label
        lbl_emotion.config(text=f"Detected Emotion: {emotion if emotion else 'No Face Detected'}")

        # Use a separate thread to avoid blocking the main loop
        if emotion:
            threading.Thread(target=speak_emotion, args=(emotion,)).start()

    # Schedule the function to be called again after 30 milliseconds
    lbl_video.after(30, update_frame)

# GUI setup
root = tk.Tk()
root.title("Emotion Detection with Webcam and TTS")

# Create video frame label
lbl_video = Label(root)
lbl_video.pack()

# Create emotion label
lbl_emotion = Label(root, text="Detected Emotion: ", font=("Helvetica", 16))
lbl_emotion.pack()

# Initialize the webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    lbl_emotion.config(text="Error: Could not open webcam.")
else:
    lbl_emotion.config(text="Webcam opened successfully.")

# Start updating frames
update_frame()

# Run the Tkinter main loop
root.mainloop()

# Release the webcam when the window is closed
cap.release()
cv2.destroyAllWindows()
```
